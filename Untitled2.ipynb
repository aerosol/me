```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
from pprint import pp

conn = sqlite3.connect("me.db")
cur = conn.cursor()
plt.close('all')
```

```python
heart_rates = pd.read_sql_query("select * from heart_rates;", conn)
steps = pd.read_sql_query("select * from steps;", conn)
```

```python
fig, ax = plt.subplots()
ax.grid(True)

steps['date'] = pd.to_datetime(steps['id'], unit='s')
steps.plot(ax=ax, x='date', y='steps', style='g-', label='Steps', figsize=(15,8), grid=True, fontsize=18)

ax3 = ax.twinx()
heart_rates['date'] = pd.to_datetime(heart_rates['id'], unit='s')
heart_rates.plot(ax=ax3, x='date', y='rate', style='r-', secondary_y=True, label='Rate', fontsize=18)
ax.set_xlabel(None)
ax.set_title("Heart rates vs steps", fontsize=18)
```
