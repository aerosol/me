```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
from pprint import pp

conn = sqlite3.connect("me.db")
cur = conn.cursor()
plt.close('all')
```

```python
heart_rates = pd.read_sql_query("select * from heart_rates where timestamp >= strftime('%s', '2020-11-19 11:00') and timestamp <= strftime('%s', '2020-11-20 12:00');", conn)
steps = pd.read_sql_query("select * from steps where id >= strftime('%s', '2020-11-19 11:00') and id <= strftime('%s', '2020-11-20 12:00');", conn)
```

```python
fig, ax = plt.subplots()
ax.grid(True)

heart_rates['date'] = pd.to_datetime(heart_rates['id'], unit='s')
heart_rates.plot(
    ax=ax, 
    x='date', 
    y='rate', 
    style='r-', 
    grid=True, 
    yticks=range(50,140,5), 
    label='Rate', 
    fontsize=18)


ax3 = ax.twinx()

steps['date'] = pd.to_datetime(steps['id'], unit='s')
steps.plot(
    ax=ax3, 
    x='date', 
    y='steps', 
    style='g-', 
    secondary_y=True, 
    label='Steps', 
    figsize=(35,20), 
    grid=True, 
    fontsize=18)


ax.set_xlabel(None)
ax.set_title("Heart rates vs steps", fontsize=18)
```
