```python
import os
import sys
import glob
from pathlib import Path
from collections import defaultdict

datasets = defaultdict(list)

# root_dir needs a trailing slash (i.e. /root/dir/)
for filename in glob.iglob('data/**/*.dat', recursive=True):
     kind = Path(filename).stem     
     datasets[kind].append(filename)
        
print(datasets)
```

```python
import json

def load_all(files):
    contents = []
    for f in files:
        with open(f) as json_file:
          contents.append(json.load(json_file))
    return contents
   
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
merged_steps = []
import itertools

def flatten(stuff):
    return list(itertools.chain(*stuff))

def concat_objects(stuff):
    return list(itertools.chain(stuff))

steps = flatten(load_all(datasets['steps']))

sleep_data = load_all(datasets['sleep'])

sleep_activity_levels = flatten([s['sleepLevels'] for s in sleep_data])

print(len(sleep_activity_levels))
sleep_activity_levels = [sal for sal in sleep_activity_levels if sal['activityLevel'] > 0.0]
print(len(sleep_activity_levels))

sleep_movements = flatten([s['sleepMovement'] for s in sleep_data])
sleep_movements = [sm for sm in sleep_movements if sm['activityLevel'] > 0.0]

respirations = flatten([s['wellnessEpochRespirationDataDTOList'] for s in sleep_data])
daily_sleep_data = concat_objects([s['dailySleepDTO'] for s in sleep_data])

heart = load_all(datasets['heart'])

heart_rates = flatten([s['heartRateValues'] for s in heart])
heart_rates = [h for h in heart_rates if h[1] is not None]
```

```python
import time, datetime
def to_unix(s):
  return int(time.mktime(datetime.datetime.strptime(s, "%Y-%m-%dT%H:%M:%S.0").timetuple()))


[ to_unix(s['startGMT']) for s in steps ]
```

```python
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100

fig, ax = plt.subplots()


ax.plot(  [ to_unix(s['endGMT']) for s in steps ], [ s['steps'] for s in steps ], label='Steps' )

ax3 = ax.twinx()
ax3.plot(  [ to_unix(s['startGMT']) for s in sleep_movements ], [ s['activityLevel'] for s in sleep_movements ], linestyle='', color='tab:green', marker='|', label='Sleep movements')

ax4 = ax.twinx()
ax4.plot(  [ to_unix(s['startGMT']) for s in sleep_activity_levels ], [ s['activityLevel'] for s in sleep_activity_levels ], label='Sleep activity', linestyle="", marker="+", color='tab:purple')

#ax5 = ax.twinx()
#ax5.plot(  [ s['startTimeGMT'] // 1000 for s in respirations ], [ s['respirationValue'] for s in respirations ], label='Respirations', color= 'tab:blue')

#ax2 = ax.twinx() 


#ax2.plot(  [ s[0] // 1000 for s in heart_rates ], [ s[1] for s in heart_rates ], label='Heart rate', color='tab:red', linestyle='', marker='.')
#ax2.tick_params(axis='y', labelcolor='tab:red')

#ax2.set_ylabel('heart rate', color= 'tab:red')
plt.show()
```
